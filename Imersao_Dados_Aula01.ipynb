##**Desafio 1:** Investigar por que a classe tratamento é tão desbalanceada?
**Resposta**:


```python
# Vamos primeiramente plotar as colunas de dados de tratamento "com_droga" e "com_controle"
sns.set(font_scale=1)
ax = dados['tratamento'].value_counts().plot(kind='bar', figsize=(4, 6))
ax.set_title('Quantidade de Tratamentos x Tipo', fontsize= 16)
ax.set_xlabel('Tipos', fontsize = 16)
ax.set_ylabel('Quantidade', fontsize=16)

#Agrupando com tipo de droga
droga_tratamento = dados.groupby(['tratamento','droga']).size()
print(droga_tratamento)
```

####Como podemos observar no gráfico acima e nos dados obtidos, o número de tratamentos "com_controle" utilizou a droga "cacb2b860", o que nos informa que talvez esse código não seja exatamente uma droga, mas a ausência da mesma, o grupo em que ocorreu a aplicação de drogas comparará resultados com com o outro em que talvez outro tipo de tratamento tenha sido feito.

##**Desafio 7:** Deixar os gráficos bonitões. (Matplotlib.pyplot)
**Resposta**:


```python
# Porcentagem por tratamento
porc_trat = dados['tratamento'].value_counts(normalize=True).reset_index()
porc_trat.columns = ['tratamento', 'proporcao']
porc_trat
```

```python
# Plotando 
fig, ax = plt.subplots(figsize=(16,6))
sns.set_theme(style='darkgrid')
sns.set_palette(sns.color_palette('pastel', 8))
sns.barplot(x='tratamento', y='proporcao', data=porc_trat, ax=ax)

# Legendas e tamanho de fontes
ax.set_xticklabels(labels=['Com Droga', 'Com Controle'])
plt.title('Proporção de cada tipo de tratamento', fontsize=18)
plt.xlabel('Tratamento', fontsize=14)
plt.ylabel('Proporção', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adicionar valores acima da barra
for index, row in porc_trat.iterrows():
  ax.text(index, row['proporcao'], round(row['proporcao']*100, 2), color='black', ha='center', fontsize=12)
plt.show()
```
