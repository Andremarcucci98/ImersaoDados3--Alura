```python
data = dados.query('composto in @cod_compostos')
data
```

```python
ordem = data['composto'].value_counts().sort_values(ascending=False)
ordem[:2][:2]
```

```python
order = data['composto'].value_counts(ascending=False).index
order
```

```python
# Configuração principal
sns.set()
plt.figure(figsize = (12,6))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order = order)
ax.set_title('5 Compostos mais utilizados', size=14, weight='bold')
ax.set_xlabel('Compostos', size=14, weight='bold')
ax.set_ylabel('Quantidade', size=14, weight='bold')

for qtd in ax.patches:
  ax.annotate(format(qtd.get_height()), (qtd.get_x() + qtd.get_width() / 2., qtd.get_height()),
               ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
plt.show()
```

Como podemos observar no gráfico de barras acima, o composto "cacb2b860" se destaca sobre os demais, logo em seguida vem "87d714366", ""9f80f3f77", etc em uma queda exponencial abrupta.

###**Desafio 04:** Estudar sobre as estatísticas retornadas no .describe()

```python
dados.describe()
```

A função .describe() gera estatísticas descritivas, estas incluem aquelas que resumem a tendência central, a dispersão e a forma da distribuição de um conjunto de dados, excluindo os valores NaN. Dentre as principais estatísticas temos as seguintes listas:

**Numeric Data**
*   count: Quantidade de dados encontrada
*   mean: Média
*   std: Desvio Padrão, distância média dos valores em relação à média
*   min: Mínimo valor encontrado 
*   max: Máximo valor encontrado
*   25%: Separa os 25% menores valores dos 75% maiores valores
*   50%: Mediana, separa os 50% menores valores dos 50% maiores valores
*   75%: Separa os 75% menores valores dos 25% maiores valores

**Object Data**
*   count: Quantidade de dados encontrada
*   unique: Retorna valores únicos
*   top: Valor mais comum
*   freq: Valor de frequência mais comum

###**Desafio 06:** Fazer outras análises com o boxplot e até com o histograma.

##Exemplo de análise usando os 5 maiores compostos utilizados e o comportamento de "g-0"

```python
#Analisando a cultura G0, como ela se comportou com os Top 05 Compostos 

sns.set()
plt.figure(figsize = (20,6))

plt.subplot(1,2,1)
ax = sns.boxplot(y='g-0',x='composto', data=dados.query('composto in @cod_compostos'), order = order)
ax.set_title('Cultura G-0 nos 5 Compostos mais usados' , fontsize= 14, weight='bold')
ax.set_xlabel('Compostos', fontsize = 14, weight='bold' )
ax.set_ylabel('g-0', fontsize=14, weight='bold')

plt.subplot(1,2,2)
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order = order)
ax.set_title('5 Compostos mais utilizados', size=14, weight='bold')
ax.set_xlabel('Compostos', size=14, weight='bold')
ax.set_ylabel('Quantidade', size=14, weight='bold')
for qtd in ax.patches:
  ax.annotate(format(qtd.get_height()), (qtd.get_x() + qtd.get_width() / 2., qtd.get_height()),
               ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
plt.show()
```

Pegando como exemplo a amostra "g-0" (poderíamos pegar qualquer outra), é observável que ocorre um maior distanciamento da média de valores se utilizamos o composto "87d714366".
