#**Aula 04 - Merge e Análise de Resultados**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()
```

```python
dados_resultados['acat_inhibitor'].unique()
```

```python
contagem_moa = dados_resultados.select_dtypes('int64').sum().sort_values(ascending=False)
contagem_moa
```

```python
dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
dados_resultados['n_moa'] != 0
```

```python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)
dados_resultados
```

```python
dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on='id')
dados_combinados.head()
```

```python
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
composto_principal = dados_combinados['composto'].value_counts().index[:5]
plt.figure(figsize=(12,8))
sns.boxplot(data = dados_combinados.query('composto in @composto_principal'), y= 'g-0', x='composto', hue='ativo_moa')
```

**concat() ->** Concatenate pandas objects along a particular axis with optional set logic along the other axes. Can also add a layer of hierarchical indexing on the concatenation axis, which may be useful if the labels are the same (or overlapping) on the passed axis number.
**pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)**

**append ->** Append rows of other to the end of caller, returning a new object.Columns in other that are not in the caller are added as new columns.
**DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=False)**

**Database-style DataFrame or named Series joining/merging**

**merge ->** merge is a function in the pandas namespace, and it is also available as a DataFrame instance method merge(), with the calling DataFrame being implicitly considered the left object in the join.
**DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)**

**join ->** DataFrame.join() is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame. 
**DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)**
